# Exploratory Data Analysis & Stacking generalize ML Pipeline

This notebook provides an exploratory review of the BiFeO3-based supercapacitor Cyclic Voltammetry (CV) dataset and demonstrates the machine learning pipeline used to predict the current curves based on physical inputs.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

# 1. Load the dataset
dataset_path = '../Data/CV_DATASET.xlsx'
if os.path.exists(dataset_path):
    df = pd.read_excel(dataset_path)
    print(f"Loaded dataset with shape: {df.shape}")
    print(df.head())
else:
    # Attempt local relative load if run in root
    dataset_path = 'Data/CV_DATASET.xlsx'
    if os.path.exists(dataset_path):
        df = pd.read_excel(dataset_path)
        print(f"Loaded dataset with shape: {df.shape}")
    else:
        print("Dataset not found. Please verify the Data/ folder contains CV_DATASET.xlsx.")
```

## 📊 Data Inspection and Summary Statistics

Let's look at the statistics of features: `Potential`, `OXIDATION`, `Zn/Co_Conc`, `SCAN_RATE`, `ZN`, and `CO`.

```python
if 'df' in locals():
    print("=== Summary Statistics ===")
    print(df.describe())
    
    print("\n=== Unique Value Profiles ===")
    print("Unique Dopant concentrations (Zn/Co_Conc):", df['Zn/Co_Conc'].unique())
    print("Unique Scan rates (SCAN_RATE):            ", df['SCAN_RATE'].unique())
    print("ZN Indicator:                              ", df['ZN'].unique())
    print("CO Indicator:                              ", df['CO'].unique())
    print("OXIDATION indicator:                       ", df['OXIDATION'].unique())
```

## 📈 Visualization of Sample CV Curves

Let's plot sample curves for Zn-doped vs Co-doped materials at different scan rates.

```python
if 'df' in locals():
    plt.figure(figsize=(10, 6))
    
    # Filter BFZO-3 (Zn/Co_Conc=3.5, ZN=1, CO=0) at 50 mV/s
    sample_df = df[(df['Zn/Co_Conc'] == 3.5) & (df['ZN'] == 1) & (df['SCAN_RATE'] == 50)]
    if not sample_df.empty:
        plt.scatter(sample_df['Potential'], sample_df['Current'], label='BFZO-3 (50 mV/s)', alpha=0.5)
        
    plt.title('Sample Cyclic Voltammetry (CV) Curve from Dataset')
    plt.xlabel('Potential (V)')
    plt.ylabel('Current (A)')
    plt.legend()
    plt.grid(True, linestyle=':', alpha=0.7)
    plt.show()
```
